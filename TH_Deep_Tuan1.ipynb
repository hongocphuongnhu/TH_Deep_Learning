```python
#Cau10: Tạo ma trận 10x10 với các phần tử ở biên có giá trị 1 và các phần tử bên trong có giá trị 0

matrix_10x10 = np.zeros((10, 10), dtype=int)

# Đặt giá trị 1 cho các phần tử ở biên
matrix_10x10[0, :] = 1
matrix_10x10[-1, :] = 1
matrix_10x10[:, 0] = 1
matrix_10x10[:, -1] = 1

print("Ma trận 10x10 với biên là 1 và bên trong là 0:")
print(matrix_10x10)
```

```python
#Cau11: Tạo ma trận 5x5 với các phần tử ở đường chéo chính là 1, 2, 3, 4, 5
diagonal_elements = np.array([1, 2, 3, 4, 5])

# Sử dụng np.diag để tạo ma trận đường chéo từ mảng 1D
matrix_diagonal = np.diag(diagonal_elements)

print("Ma trận 5x5 với đường chéo chính là 1, 2, 3, 4, 5:")
print(matrix_diagonal)
```

```python
#Cau12: Tạo mảng 3x3x3 và tính tổng theo mỗi dòng, mỗi cột

my_3d_array = np.random.randint(0, 21, size=(3, 3, 3))
print("Mảng 3x3x3:")
print(my_3d_array)

sum_along_rows = np.sum(my_3d_array, axis=2)
print("\nTổng theo 'dòng' (axis=2, tổng các cột trong mỗi 3x3 slice):")
print(sum_along_rows)

sum_along_columns = np.sum(my_3d_array, axis=1)
print("\nTổng theo 'cột' (axis=1, tổng các hàng trong mỗi 3x3 slice):")
print(sum_along_columns)

sum_along_depth = np.sum(my_3d_array, axis=0)
print("\nTổng theo 'sâu' (axis=0, tổng qua các slice 3x3):")
print(sum_along_depth)
```

```python
#Cau15
import pandas as pd

series1 = pd.Series([2, 4, 6, 8, 10])
series2 = pd.Series([1, 3, 5, 7, 10])

print("Series 1:")
print(series1)
print("Series 2:")
print(series2)

cong_result = series1 + series2
print("\nKết quả Cộng:")
print(cong_result)

tru_result = series1 - series2
print("Kết quả Trừ:")
print(tru_result)

nhan_result = series1 * series2
print("Kết quả Nhân:")
print(nhan_result)

chia_result = series1 / series2
print("Kết quả Chia:")
print(chia_result)
```

```python
#Cau17
data = {
    'Cột X': np.random.randint(1, 100, size=4),
    'Cột Y': np.random.randint(1, 100, size=4),
    'Cột Z': np.random.randint(1, 100, size=4)
}
df_random = pd.DataFrame(data)

print("DataFrame gốc:")
print(df_random)

df_sorted_by_X = df_random.sort_values(by='Cột X')
print("\nDataFrame sau khi sắp xếp theo 'Cột X':")
print(df_sorted_by_X)

#giam dan
df_sorted_by_Y_desc = df_random.sort_values(by='Cột Y', ascending=False)
print("\nDataFrame sau khi sắp xếp theo 'Cột Y' (giảm dần):")
print(df_sorted_by_Y_desc)
```
